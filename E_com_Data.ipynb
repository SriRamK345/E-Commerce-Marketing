```python
import docx
import pandas as pd

# Function to extract text from a .docx file
def read_docx(file_path):
    doc = docx.Document(file_path)
    doc_text = []
    for para in doc.paragraphs:
        doc_text.append(para.text)
    return "\n".join(doc_text)

# Specify the path to the .docx file
file_path = "/content/drive/MyDrive/E-Commerce Marketing/Product Details.docx"

# Extract text from the .docx file
doc_text = read_docx(file_path)

# Optionally, split the text into a list (e.g., by paragraphs or lines)
doc_lines = doc_text.split('\n')  # Split the string into lines

# Load into a pandas DataFrame for further processing
# Create a list of lists, where each inner list represents a row
# with a single column containing the text from each line
product_details = pd.DataFrame([ [line] for line in doc_lines], columns=['Text'])

# Display the DataFrame
product_details.head()
```

```python
import docx
import pandas as pd

# Function to extract tables from a .docx file
def extract_tables_from_docx(file_path):
    doc = docx.Document(file_path)
    all_tables = []

    for table in doc.tables:
        table_data = []
        for row in table.rows:
            row_data = [cell.text for cell in row.cells]
            table_data.append(row_data)
        all_tables.append(pd.DataFrame(table_data))

    return all_tables

file_path = "/content/drive/MyDrive/E-Commerce Marketing/Product Details.docx"
# Extract tables from the .docx file
tables = extract_tables_from_docx(file_path)
product_details = pd.DataFrame(tables[0])
product_details.head()
```

```python
product_details.columns = product_details.iloc[0]
product_details = product_details.drop(0)
product_details.reset_index(drop=True, inplace=True)
```
